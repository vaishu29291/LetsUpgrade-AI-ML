```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ds=pd.read_csv('general_data.csv')
```

# Wilcoxon Test

```python
from scipy.stats import wilcoxon
stats,p=wilcoxon(ds.JobLevel,ds.Education)
print(stats,p)
```

```python
from scipy.stats import friedmanchisquare
stats,p=friedmanchisquare(ds.StandardHours,ds.StockOptionLevel,ds.TotalWorkingYears)
print(stats,p)
```

# Mannwhiteneyu Test

```python
from scipy.stats import mannwhitneyu
stats,p=mannwhitneyu(ds.DistanceFromHome,ds.Education)
print(stats,p)
```

```python
2.371790832392393e-02
```

```python
from scipy.stats import ttest_1samp
stats,p=ttest_1samp(ds.JobLevel,76)
print(stats,p)
```

# Two Sample Paired T-test

```python
from scipy.stats import ttest_rel
stats,p=ttest_rel(ds.JobLevel,ds.Education)
print(stats,p)
```

```python
from sklearn import preprocessing
Label_encoder=preprocessing.LabelEncoder()
ds['Gender']=Label_encoder.fit_transform(ds['Gender'])
ds['MaritalStatus']=Label_encoder.fit_transform(ds['MaritalStatus'])
```

```python
ds.head()
```

# Chi square Test

```python
from scipy.stats import chi2_contingency


chitable = pd.crosstab(ds.Gender, ds.MaritalStatus)
stats, p, dof, expected = chi2_contingency(chitable)
print(stats,p)
```
